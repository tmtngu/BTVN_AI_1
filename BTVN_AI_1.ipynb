```python
import folium
uehB = [10.7613, 106.6675]
m = folium.Map(location=uehB, zoom_start=15)

folium.Marker(
    location=uehB,
    popup="<b>Đại học UEH - Cơ sở B</b>",
    tooltip="Trường Đại học",
    icon=folium.Icon(color='blue', icon='graduation-cap', prefix='fa')
).add_to(m)

congcong = folium.FeatureGroup(name='Địa điểm công cộng').add_to(m)

locations = [
    {"name": "Bệnh viện Chợ Rẫy", "coord": [10.7578, 106.6595], "info": "Bệnh viện đa khoa lớn nhất"},
    {"name": "Bệnh viện Đại học Y Dược", "coord": [10.7555, 106.6653], "info": "Cơ sở 1"},
    {"name": "Sân vận động Thống Nhất", "coord": [10.7616, 106.6610], "info": "Sân vận động trung tâm"},
    {"name": "The Garden Mall", "coord": [10.7548, 106.6625], "info": "Trung tâm thương mại giải trí"},
    {"name": "Công viên Hòa Bình", "coord": [10.7585, 106.6685], "info": "Khuôn viên xanh gần trường"}
]

for loc in locations:
    folium.Marker(
        location=loc["coord"],
        popup=f"<b>{loc['name']}</b><br>{loc['info']}",
        tooltip=loc['name'],
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(congcong)

folium.LayerControl().add_to(m)

m
```

```python
import folium
from geopy.distance import geodesic
import time

uehB = [10.7613, 106.6675]
m = folium.Map(location=uehB, zoom_start=14)

folium.Marker(
    uehB,
    popup="<b>ĐIỂM GỐC: UEH Cơ sở B</b>",
    icon=folium.Icon(color='blue', icon='star', prefix='fa')
).add_to(m)

ueh_campuses = [
    {"name": "Cơ sở A", "addr": "59C Nguyễn Đình Chiểu, Q.3", "coord": [10.7801, 106.6953]},
    {"name": "Cơ sở C", "addr": "91 Ba Tháng Hai, Q.10", "coord": [10.7707, 106.6789]},
    {"name": "Cơ sở D", "addr": "196 Trần Quang Khải, Q.1", "coord": [10.7915, 106.6897]},
    {"name": "Cơ sở E", "addr": "54 Nguyễn Văn Thủ, Q.1", "coord": [10.7872, 106.6983]},
    {"name": "Cơ sở H", "addr": "1A Hoàng Diệu, Q. Phú Nhuận", "coord": [10.7997, 106.6755]},
    {"name": "Cơ sở I", "addr": "17 Phạm Ngọc Thạch, Q.3", "coord": [10.7828, 106.6961]},
    {"name": "Cơ sở N", "addr": "Nam Thành phố (Bình Chánh)", "coord": [10.7011, 106.6432]},
    {"name": "Cơ sở V", "addr": "232/6 Võ Thị Sáu, Q.3", "coord": [10.7841, 106.6872]},
    {"name": "Ký túc xá 135 Trần Hưng Đạo", "addr": "135 Trần Hưng Đạo, Q.1", "coord": [10.7655, 106.6934]},
    {"name": "Ký túc xá 43-45 Nguyễn Chí Thanh", "addr": "43-45 Nguyễn Chí Thanh, Q.5", "coord": [10.7597, 106.6713]}
]

print("Đang tính toán khoảng cách giữa các cơ sở UEH ...")

for campus in ueh_campuses:
    dist = geodesic(uehB, campus["coord"]).km

    folium.Marker(
        location=campus["coord"],
        popup=f"<b>{campus['name']}</b><br>Địa chỉ: {campus['addr']}<br>Cách cơ sở B: {dist:.2f} km",
        tooltip=campus['name'],
        icon=folium.Icon(color='red', icon='location-dot', prefix='fa')
    ).add_to(m)

    folium.PolyLine(
        [uehB, campus["coord"]],
        color="black",
        weight=1,
        opacity=0.6,
        dash_array='5'
    ).add_to(m)

m
```

# CÂU 3

```python
import folium
from folium.plugins import HeatMap
import numpy as np

uehB = [10.7613, 106.6675]
m = folium.Map(location=uehB, zoom_start=14)

np.random.seed(42)
data = (np.random.normal(size=(100, 2)) * 0.008 + uehB).tolist()

HeatMap(data, radius=15, blur=10, min_opacity=0.5).add_to(m)

folium.Marker(
    uehB,
    popup="UEH csB",
    icon=folium.Icon(color='red', icon='star', prefix='fa')
).add_to(m)

m
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

try:
    vietnam_map = gpd.read_file('/content/vn_geo.json')
    print("Đã tải bản đồ thành công!")
except Exception as e:
    print(f"Lỗi không tìm thấy file: {e}. Bạn hãy đảm bảo đã upload file vn_geo.json vào mục Files.")

data = {
    'name': [
        'Hồ Chí Minh', 'Hà Nội', 'Đà Nẵng', 'Bình Dương', 'Đồng Nai',
        'Hải Phòng', 'Cần Thơ', 'Bà Rịa - Vũng Tàu', 'Quảng Ninh', 'Long An'
    ],
    'Revenue': [5200, 4800, 2100, 3900, 3400, 2800, 1900, 2500, 2200, 1600],
    'Order_Volume': [15000, 14000, 6000, 11000, 10500, 8000, 5500, 7000, 6500, 4000]
}
df_business = pd.DataFrame(data)

merged_data = vietnam_map.merge(df_business, on='name', how='left').fillna(0)

fig, ax = plt.subplots(1, 1, figsize=(10, 15))

merged_data.plot(
    column='Revenue',
    cmap='YlOrRd',
    ax=ax,
    edgecolor='0.4',
    legend=True,
    legend_kwds={'label': "Quy mô doanh thu (Tỷ VNĐ)", 'orientation': "horizontal"}
)
ax.set_title('BẢN ĐỒ PHÂN TÍCH DOANH THU THEO ĐƠN VỊ HÀNH CHÍNH', fontsize=16, fontweight='bold')
ax.axis('off')
plt.figtext(0.15, 0.1,
            "Nhận xét quản trị:\n"
            "- Các vùng đỏ đậm tập trung tại 2 cực kinh tế và vùng vệ tinh (Bình Dương, Đồng Nai).\n"
            "- Cần tập trung nguồn lực logistics vào cụm phía Nam để tối ưu chi phí vận hành.",
            ha="left", fontsize=10, style='italic', bbox={'facecolor':'orange', 'alpha':0.1, 'pad':5})

plt.show()
```
